```python
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head()
```

```python
# nombres des lignes et des colonnes
Observation = df.shape[0]
Valeurs = df.shape[1]
print("Observation: ",Observation)
print("Valeurs:", Valeurs)
```

```python
#  data types of each columns
```

```python
df.dtypes
```

```python
# cleanup data
```

```python
# we gonna combine 50km/50mi of 2020 for the USA
```

```python
df[(df["Event distance/length"].isin(["50mi","50km"])) & (df["Year of event"] == 2020) ]
```

```python
df[df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA" ]
```

```python
# combine all the filters togethers (USA,2020,50mi/50km)
```

```python
df2=df[(df["Event distance/length"].isin(["50mi","50km"])) & (df["Year of event"] == 2020) & ( df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA" )]
```

```python
df2.head()
```

```python
# the numbers of the new filtered data
Observation = df2.shape[0]
Valeurs = df2.shape[1]
print("Observation: ",Observation)
print("Valeurs:", Valeurs)
```

```python
# remove the state from the event name
```

```python
df2["Event name"]=df2["Event name"].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# clean up the age of the athlete
df2["Athlete_age"] = 2020-df2["Athlete year of birth"]
```

```python
df2.head()
```

```python
# remove h for the athelete performance
```

```python
df2["Athlete performance"]= df2["Athlete performance"].str.split(" ").str.get(0)
```

```python
df2.head()
```

```python
df2.drop(labels=["Athlete club","Athlete country", "Athlete year of birth","Athlete age category"],axis=1,inplace=True)
```

```python
df2.shape
```

```python
df2.head()
```

```python
#clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2["Athlete_age"].isna()==True]
```

```python
df2.dropna(inplace=True)
```

```python
df2.shape
```

```python
# check for duplicate values
```

```python
df2[df2.duplicated()==True]
```

```python
#reset index
```

```python
df2.reset_index(inplace=True)
```

```python
#fix types
```

```python
df2.dtypes
```

```python
df2["Athlete_age"] = df2["Athlete_age"].astype(int)
```

```python
df2["Athlete average speed"] = df2["Athlete average speed"].astype(float)
```

```python
# rename columns
```

```python
# Year of event                  
# Event dates                   
# Event name                    
# Event distance/length         
# Event number of finishers      
# Athlete performance           
# Athlete gender                
# Athlete average speed        
# Athlete ID                    
# Athlete_age                
```

```python
df2=df2.rename(columns={"Year of event":"year"     ,             
"Event dates":"race_day",                
"Event name":"race_name"  ,                  
"Event distance/length":"race_length",         
"Event number of finishers":"race_number_of_finishers",      
"Athlete performance":"athlete_performance"   ,        
"Athlete gender":"athlete_gender"         ,       
"Athlete average speed":"athlete_average_speed"   ,     
"Athlete ID":"athlete_id",
 "Athlete_age " :"athlete_age"  
})
```

```python
#reorder clolumns
df2.dtypes
```

```python
df3=df2[["race_day","race_name","race_length","race_number_of_finishers","athlete_id","athlete_gender","Athlete_age","athlete_performance","athlete_average_speed","year"]]
```

```python
#Seasons for the data are they slower in summer than winter ?
#spring 3-5
#summer 6-8
#fall 9-10
#winter 12-2

#split between two decimals
```

```python
df3['race_moth']=df3["race_day"].str.split(".").str.get(1).astype(int)
```

```python
df3
```

```python
df3.dtypes
```

```python
df3["race_season"]= df3["race_moth"].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3
```

```python
df3.groupby("race_season")["athlete_average_speed"].agg(["mean","count"]).sort_values("mean",ascending=False)
```
